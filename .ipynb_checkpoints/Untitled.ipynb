```python
# Code contributors: Niharika Rajwadha, Jeevash Mutreja, Chetas Das
import random 
import time
import datetime 
import collections
import dateutil
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
import matplotlib.axes as axes
import pickle, gzip
import os
from random import randrange
import ipywidgets as widgets
from ipywidgets import interact
%matplotlib inline

class person:

    def __init__(self):
        self.name="Test Subject";
        self.age = 0
        self.gender = "M"
        self.isAthletic = False
        
    def __str__():
        print("This represents person class. This contains patients' personal information")

    def setAttributes(self, name, age, gender, isAthletic):
        self.name = name
        self.age = age
        self.gender = gender
        self.isAthletic = isAthletic

    def getAttributes(self):
        #print personal data
        print ("Name: ", self.name, ", Age: ", self.age, ", Gender: ", self.gender, ", Is Athlete: ", self.isAthletic)
        
    def getName(self):
        return self.name

class health(person):
    
    def __init__(self):
        self.cardiacDisease = False
        self.diseaseType = "None"
        self.wasInPain = False

    def __str__():
        print("This represents person's health class. This contains patients' personal health related information")
        
    def setHealthAttributes(self, cardiacDisease, diseaseType, wasInPain):
        self.cardiacDisease = cardiacDisease
        self.diseaseType = diseaseType
        self.wasInPain = wasInPain

    def getHealthAttributes(self):
        #print details
        print("Cardiac Disease: ", self.cardiacDisease, ", Disease Type: ", self.diseaseType, ", Experiencing pain in body: ", self.wasInPain)

class plot(health):
        
    def __str__():
        print("This represents plot class. This contains information for plotting the various graphs/visualizations")
    
    def getData(self, start, l, diseased=False):
        current = start
        while l >= 0:
            current = current + datetime.timedelta(seconds=random.gauss(1,0.55))
            #if current.date() == start.date():
            yield current #if yes, it returns the result and goes back to the loop
            l-=1 #decrement the value of the timestamp counter
    
    #method END
    
    def readData(self):
        #some code to read from file
        with gzip.open("dd_dict","rb") as fo:
            dict_19_12 = pickle.load(fo)

    def writeData(self, pickle_dict,write_to='pickled_data',write_over=False):
        #some code to write data into file
        if os.path.exists(write_to) and (write_over == False):
            raise FileExistsError("File with that name already exists")
        else:
            f = gzip.open(write_to,'wb')
            pickle.dump(pickle_dict,f)
            f.close()
    
    def showPlots(self, beats):
        #some code to show the plots of heart rate
        df=pd.DataFrame(beats)
        df.columns=(['Date','Hour','Minute','Second'])
        df.describe()
        df2= df.groupby(['Date','Hour'])
        
        #Calculating Averager Beat in a day for everyday
        Average_Beats=pd.DataFrame((df2.count()/60)['Minute'].reset_index())
        print(Average_Beats)
        
        #Average Beats per minute per hour in a day
        sns.set(font_scale=1) 
        g= sns.FacetGrid(data= Average_Beats,row='Date',ylim=(49,65), xlim=(0,24), aspect= 2) 
        g.map(sns.pointplot, 'Hour','Minute')
        
        #Box Plot to show Maximum Minimun and Average for different days
        sns.set(font_scale=3)
        plt.figure(figsize= (40,15))
        plt.ylim(47,65)
        sns.boxplot(x= 'Date', y='Minute',data=Average_Beats)
        return Average_Beats

#INPUT PATIENT DETAILS
patient = person()
condition = health()
graph = plot()

patient.setAttributes("Mike", 30, "M", False)
patient.getAttributes()

condition.setHealthAttributes(True, "Aortic Stenosis", False)
condition.getHealthAttributes()

#START ANALYSIS PLOTTING
#here getData function can be used to get the data from the devices connected to the heart and then use the collected data to plot the graphs of heart's performance
startDate = datetime.datetime(2017, 12, 19,23,59,00) #seed value 
dd = collections.defaultdict(list)
for x in (list(graph.getData(startDate,1000000, True))): 
    day = (x.strftime("%d/%m/%y")) 
    time = (x.strftime("%H:%M:%S")) 
    dd[day].append(time)
```

```python
graph.writeData(dd,patient.getName(),True)
graph.readData()

def view(x=''):
    if x=='All':
        return Average_Beats
    else:
        return Average_Beats[Average_Beats['Date']==x]

#spliting time into Hours, Minutes and Seconds
beats=[] 
for k in dd: 
    time=dd[k] 
    for w in time: 
        striplist = w.strip().split(':') 
        beats.append((k,striplist[0],striplist[1],striplist[2]))
Average_Beats = graph.showPlots(beats)

#Created Interactive Widget showing date wise average beats
items = ['All']+sorted(Average_Beats['Date'].unique().tolist())
w = widgets.Select(options=items)
interact(view, x=w)
```
